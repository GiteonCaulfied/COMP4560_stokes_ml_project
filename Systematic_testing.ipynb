```python
import Systematic_Model_Geoid_1D as model
```

```python
# Using readlines()
file1 = open('ModelList.txt', 'r')
Lines = file1.readlines()
 
count = 0
# Strips the newline character
for line in Lines:
    if line[0] != '#':
        
        hidden_layers_string, act_function, loss_function, n_epoch, batch_size, lr = line.strip().split(',')
        hidden_layers_list = [int(x) for x in hidden_layers_string.split('/')]
        
        count += 1
        print("Model {}: ".format(count))
        print("Hidden Layers Nuerons: ", hidden_layers_string)
        print("Activation Function: ", act_function)
        print("Loss Function: ", loss_function)
        print("Total epochs: ", n_epoch)
        print("Batch size: ", batch_size)
        print("Optimizer Learning rate: ", lr)
        print()
        
        model.run_model(hidden_layers_list, act_function, loss_function, int(n_epoch), int(batch_size), float(lr), count)
        print()
```
